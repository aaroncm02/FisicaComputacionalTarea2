```python
import numpy as np
import matplotlib.pyplot as plt

def gaussxw(N):
    x, w = np.polynomial.legendre.leggauss(N)
    return x, w

def gaussxwab(a, b, x, w):
    xp = 0.5 * (b - a) * x + 0.5 * (b + a)
    wp = 0.5 * (b - a) * w
    return xp, wp

def integrando(x):
    return np.sin(x**2)

a, b = 0, np.pi
Ns = list(range(2, 51))
resultados = []

for N in Ns:
    x, w = gaussxw(N)
    xp, wp = gaussxwab(a, b, x, w)
    I = np.sum(wp * integrando(xp))
    resultados.append(I)

plt.plot(Ns, resultados, 'o-', label=r'$\int_0^\pi \sin(x^2)\,dx$')
plt.xlabel('N')
plt.ylabel('Resultado de la integral')
plt.title('Convergencia de la Cuadratura Gaussiana')
plt.grid(True)
plt.legend()
plt.show()

print(f"Valor estimado de la integral con N=50: {resultados[-1]:.10f}")

```
